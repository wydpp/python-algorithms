## 1.问题描述

有一对兔子，从出生后的第3个月起，每个月都会生一对兔子。
小兔子长到第3个月后每个月又生一对兔子。
假设所有的兔子都不死，问30个月内每个月的兔子总对数为多少？

## 2.问题分析

画一张表来找一下兔子数的规律

| 月数   | 小兔子对数 | 中兔子对数   |老兔子对数 |
|:--------:|:------:|:------------:|:------------:|
| 1   | 1   | 0     | 0  |1|
| 2   | 0   | 1     |  0 |1|
| 3   | 1   | 0     |  1 |2|
| 4   | 1   | 1     |  1 |3|
| 5   | 2   | 1     |  2 |5|
| 6   | 3   | 2     |  3 |8|

说明：不满一个月的为小兔子，满1个月不满2个月的为中兔子，满3个月以上的为老兔子。老兔子每个月都会生一对兔子。

可以看出，每个月兔子的总数依次为1，1，2，3，5，8...这就是Fibonacci数列。

总结规律即为前两个月的兔子对数可以退出第三个月的兔子对数。

## 3.算法设计

本题目是典型的迭代循环，即是一个不断用新值取代变量的旧值，然后由变量旧值递推出变量新值的过程。

这种迭代与这些因素有关：初值、迭代公式和迭代次数。算法可以描述为：

$$
\begin{cases} 
f(2) = f(1) = 1 & (n < 3) &\text{初值}\\
f(n) = f(n-1) + f(n-2) & (n \geq 3) &\text{迭代公式} 
\end{cases}
$$

## 4.完整程序

```python


def cal1():
    f1 = 1
    f2 = 1
    #月份
    i = 3
    while(i <=30):
      f3 = f1 + f2
      print(f"第{i}个月的兔子总对数是：{f3}")
      f1 = f2
      f2 = f3
      i += 1
cal1()
```

## 5.算法优化

可以每次循环计算出f1和f2的值，这样可以循环15次就可以得到结果。

```python
def cal2():
    f1 = 1
    f2 = 1
    i = 2
    while i <= 15:
        # 每次求两个值
        f1 = f1 + f2
        print(f"第{i*2-1}个月的兔子总对数是：{f1}")
        f2 = f1 + f2
        print(f"第{i*2}个月的兔子总对数是：{f2}")
        i += 1
cal2()
```

## 6.递归算法

```python
def cal3(n):
    if(n < 3):
        return 1
    return cal3(n-1) + cal3(n-2)
print(f"第30个月的兔子数量：{cal3(30)}")
```
